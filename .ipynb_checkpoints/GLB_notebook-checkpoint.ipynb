```python
import pandas as pd

pd.set_option("max_colwidth", 500)
pd.options.display.max_rows = 999
pd.options.display.max_columns = 100
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# laad .zip
subsidies_file = "data/GLB_2014_2021.zip"

df = pd.read_csv(subsidies_file, compression='zip', 
                 dtype={"jaar" : "str"})

print(f'{len(df)} rows loaded')
```

## Beschrijving van de data

|*kolom*|*ontbreekt*|*datatype*|*beschrijving*|
|---|---|---|---|
|**index**|0 ontbreekt | object | ID van de betaling|
|**bedrijf_ID**|0 ontbreekt | object | ID gemaakt op basis van de originele gegevens, in het formaat `postcode-bedrijfsnaam-plaats`|
|**bedrijf**|0 ontbreekt | object | bedrijfsnaam zoals die voorkomt in de originele gegevens|
|**plaats**|0.01% ontbreekt | voorwerp| plaats zoals die voorkomt in de originele gegevens.<br> **NB**: Alleen ter referentie. Dit kan het dorp, de stad of gemeente zijn, maar ook de gemeente. Gebruik voor analyse `woonplaats` of `gemeente`|
|**postcode**|0.01% ontbreekt | object | postcode afkomsting uit de originele gegevens. Postcode kan in een ander land zijn dan NL - BE of DE, zelfs IE|
|**categorie**|0 ontbreekt | object | categorienummer van de subsidie. Opgeschoond voor gemengde subsidies / typfouten|
|**categorie_naam**|0 ontbreekt | object | categorienaam in NL. Opgeschoond <br> (zie ook `Omschrijving_maatregelen.docx`)|
|**bedrag**|0 ontbreekt|float| betaald bedrag|
|**jaar**|0 ontbreekt |object|jaar waarin de subsidie is verleend|
|**woonplaats**|0,67% vermist|object|plaats volgens de postcode|
|**gemeente**|0,71% vermist |object|gemeente volgens de postcode|
|**provincie**|0,67% ontbreekt |object|provincie volgens de postcode|
|**autoriteit**|0 ontbreekt |object|autoriteit die de subsidie verstrekt|
